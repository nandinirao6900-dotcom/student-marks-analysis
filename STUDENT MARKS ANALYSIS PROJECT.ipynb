```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = {
    "names": ["nandini","shivam", "megha"],
    "maths"  :[78,89,90],
    "english":[67,90,94],
    "science":[23,67,89],

        
}
```

```python
df = pd.DataFrame(data)
```

```python
df["total"] = df["maths"] + df["english"] + df["science"]
```

```python
df["percentage"] = df["total"] / 3
```

```python
print("THE STUDENT DATA IS : ")
```

```python
df
```

```python
average_percentage = np.mean(df["percentage"])
highest_percentage = np.max(df["percentage"])
lowest_percentage  = np.min(df["percentage"])
```

```python
print("average percentage = ", round(average_percentage,2))
print("highest percentage = ", round(highest_percentage,2))
print("lowest percentage = ", round(lowest_percentage,2))
```

```python
df.to_csv("studentmarks.csv",index = False)
```

```python
plt.figure(figsize = (8, 5))
plt.bar(df["names"], df["percentage"])
plt.title("student percentage analysis")
plt.xlabel("names")
plt.ylabel("marks")
plt.show()
```

```python
subject_avg = [
    df["maths"].mean(),
    df["english"].mean(),
    df["science"].mean(),
]

subjects = ["Maths", "English", "Science"]

plt.figure(figsize=(6, 6))
plt.pie(subject_avg, labels=subjects, autopct="%1.1f%%")
plt.title("Subject Wise Average")
plt.show()
```
